```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Importar bibliotecas para a rede neural
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Importar bibliotecas para o pré-processamento

import os

os.chdir(os.path.dirname(os.getcwd())) # Mudar o diretório de trabalho para a raiz do projeto
print("diretório:", os.getcwd())


dados = np.load('../dados/dados_processados.npz') # Carregar os dados processados (salvos como arquivo .npz)

X_train = dados['X_train'] # Separar os dados de treino, validação e teste
y_train = dados['y_train']
X_val = dados['X_val']
y_val = dados['y_val']
X_test = dados['X_test']
y_test = dados['y_test']
```

```python
# Definindo arquitetura do modelo, neste caso utilizando tensorflow.keras. Com 2 camadas ocultas e dropout para regularização.
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'), # Camada de entrada com 64 neurônios e função de ativação ReLU
    Dropout(0.5), # Dropout para regularização
    Dense(32, activation='relu'), # Camada oculta com 32 neurônios e função de ativação ReLU
    Dropout(0.5), # Dropout para regularização
    Dense(1, activation='sigmoid')  # Camada de saída com 1 neurônio e função de ativação Sigmoid
])

# Resumo do modelo
model.summary()
```

```python
# Compilar o modelo
model.compile(optimizer='adam', # Otimizador Adam
              loss='binary_crossentropy', # Função de perda Binary Crossentropy
              metrics=['accuracy']) # Métrica de avaliação Acurácia

# Early Stopping para evitar overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True) # Parar o treinamento se a perda de validação não melhorar por 10 épocas

# Treinar o modelo
history = model.fit(X_train, y_train, # Dados de treino
                    epochs=100, # Número de épocas
                    batch_size=32, # Tamanho do batch
                    validation_data=(X_val, y_val), # Dados de validação
                    callbacks=[early_stop], # Early Stopping
                    verbose=1) # Exibir progresso do treinamento

# em resumo o modelo foi treinado com 100 épocas, batch size de 32 e validação a cada época. O treinamento foi interrompido se a perda de validação não melhorou por 10 épocas.
```

```python
# Avaliar o modelo no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0) # Avaliar o modelo no conjunto de teste
print(f'Precisão no Teste: {test_accuracy*100:.2f}%') # Exibir a precisão no conjunto de teste

# em resumo, o trecho acima avalia o modelo treinado no conjunto de teste e exibe a precisão obtida.
```

```python
# Verirficar apenas a precisão por si só pode ser enganosa, 
# especialmente se houver um desequilíbrio nas classes (ou seja, se uma classe for significativamente mais frequente que a outra).
# Por isso, iremos também verificar a distribuição das classes no conjunto de teste.
# Com isso iremos ver se o modelo está aprendendo a prever apenas a classe majoritória.
import numpy as np

unique, counts = np.unique(y_test, return_counts=True)
class_distribution = dict(zip(unique, counts))
print("Distribuição das Classes no Teste:", class_distribution)
```

As classes estão balanceadas.

72 músicas para "Não Popular" e 71 para "Popular".

```python
# Previsões no conjunto de teste
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int).flatten()

# Relatório de Classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))

# Matriz de Confusão
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Popular', 'Popular'], yticklabels=['Não Popular', 'Popular'])
plt.ylabel('Verdadeiro')
plt.xlabel('Predito')
plt.title('Matriz de Confusão')
plt.show()

# em resumo, o trecho acima exibe o relatório de classificação e a matriz de confusão para avaliar o desempenho do modelo no conjunto de teste.
```

- **F1-Score**: 0,61

#### Classe 1 (Popular)

- **F1-Score**: 0,55

### Matriz de Confusão


- **Verdadeiros Negativos (46)**: Músicas que eram realmente "Não Popular" e foram corretamente classificadas como tal.
- **Falsos Positivos (26)**: Músicas que eram "Não Popular" mas foram incorretamente classificadas como "Popular".

- **Verdadeiros Positivos (37)**: Músicas que eram realmente "Popular" e foram corretamente classificadas como tal.


```python
# Plotar as curvas de perda
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Perda Treino')
plt.plot(history.history['val_loss'], label='Perda Validação')
plt.title('Perda durante o Treinamento')
plt.xlabel('Época')
plt.ylabel('Perda')
plt.legend()

# Plotar as curvas de precisão
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Precisão Treino')
plt.plot(history.history['val_accuracy'], label='Precisão Validação')
plt.title('Precisão durante o Treinamento')
plt.xlabel('Época')
plt.ylabel('Precisão')
plt.legend()

plt.tight_layout()
plt.show()

# em resumo, o trecho acima exibe o relatório de classificação, a matriz de confusão e as curvas de perda e precisão durante o treinamento do modelo.
```

Embora as classes estejam balanceadas, o modelo mostra um desempenho inconsistente entre as classes, especialmente no recall da classe "popular".

A classe "Não Popular" tem um recall de (64%), alto em comparação com a classe "Popular" (52%). Sugerindo que o modelo é melhor em identificar músicas que não são populares do que aquelas que são populares.

### Recall

Recall é uma métrica de avaliação que mede a capacidade do modelo em identificar corretamente todas as instâncias positivas de uma classe. 

Um recall alto indica que o modelo está acertando a maioria dos casos positivos. 

Um recall baixo sugere que o modelo está perdendo muitos casos que deveriam ser identificados como positivos.

### Importância do Recall

No contexto deste modelo, o objetivo principal é identificar músicas populares. Um recall baixo para a classe "Popular" significa que o modelo está classificando muitas músicas populares como não populares (gerando muitos falsos negativos).

## Foco nas Métricas de Recall

Dado que o objetivo é identificar músicas populares, precisamos aumentar o recall para essa classe, minimizando os falsos negativos.
