```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar o estilo dos gráficos
sns.set(style='whitegrid')
%matplotlib inline
```

```python
import os # Biblioteca para manipulação de arquivos
print("Diretório de trabalho atual:", os.getcwd()) # Mostra o diretório de trabalho atual
```

```python
# Ler o arquivo CSV com os parâmetros adequados

import os
# Mudar para o diretório pai de .ipynb_checkpoints (que é notebooks)
os.chdir(os.path.dirname(os.getcwd()))
print("diretório:", os.getcwd())


df = pd.read_csv('../dados/Spotify_Most_Streamed_Songs.csv', # define parâmetros de leitura do arquivo CSV.
                 sep=',',            # Separador é a vírgula
                 quotechar='"',      
                 engine='python',   
                 skipinitialspace=True,  # Ignora espaços após o separador
                 na_values=['', ' '],    # Trata strings vazias como NaN
                 on_bad_lines='skip'     # Ignora linhas com erros
                 )

# Exibir as primeiras linhas de dado
print(df.head())
```

```python
df.head() # Exibir as primeiras linhas de dado
```

```python
# Exibir informações gerais sobre a base de dados
print(df.info())

```

```python
print(df['streams'].dtype) # Exibir o tipo de dado da coluna 'streams'
```

```python
## converter a coluna streams para tipo adequado para manipulação

df['streams'] = df['streams'].astype(str) # converter para string
df['streams'] = df['streams'].replace('nan', np.nan) # substituir 'nan' por NaN (Stringn para NaN)
df.dropna(subset=['streams'], inplace=True) # remover linhas com NaN
```

```python
# Remover vírgulas dos valores de streams e converter dados de streams para dados numérico
df['streams'] = df['streams'].str.replace(',', '') # remover vírgulas
df['streams'] = pd.to_numeric(df['streams'], errors='coerce') # converter para numérico


# Remover linhas com valores nulos em 'streams'
df.dropna(subset=['streams'], inplace=True)

# Verificar valores nulos em 'streams'
print("Valores nulos em 'streams':", df['streams'].isnull().sum())
```

```python
print(df[['streams']].head()) # Exibir as primeiras linhas da coluna 'streams'
```

```python
print("Tipo de dados após a conversão:", df['streams'].dtype) # Verificar o tipo de dados da coluna 'streams'
```

```python
# distribuição dos streams

plt.figure(figsize=(10,6))
sns.histplot(df['streams'], bins=50, kde=True)
plt.title('Distribuição dos Streams')
plt.xlabel('Número de Streams')
plt.ylabel('Frequência')
plt.show()
```

```python
# top 10 músicas com mais streams

top10 = df.sort_values('streams', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='streams', y='track_name', data=top10, palette='viridis')
plt.title('Top 10 Músicas com Mais Streams')
plt.xlabel('Número de Streams')
plt.ylabel('Música')
plt.show()
# O gráfico mostra as 10 músicas mais ouvidas no Spotify
```

```python
df.describe() # Exibir estatísticas descritivas
```

```python
# seleciona apenas colunas numericas
df_numerico = df.select_dtypes(include=['float64', 'int64'])

# calcula matriz de correlação
matriz_correlacao = df_numerico.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm')
plt.title('Mapa de calor entre variáveis numéricas')
plt.show()
# A matriz de correlação mostra a relação entre as variáveis numéricas do dataset.
```
